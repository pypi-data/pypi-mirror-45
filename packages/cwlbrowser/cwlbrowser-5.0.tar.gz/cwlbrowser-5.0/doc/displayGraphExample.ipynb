## Graphical view of workflows
Calling the function `displayGraph(workflow)` will display a graphical representation of a workflow. The attribute 'workflow' is the reference to the `Workflow` object.

```python
import cwlbrowser.browser as b
textures = b.load('https://github.com/undu/stellaris-emblem-lab/blob/cwl/textures/emblem_textures.cwl')
b.displayGraph(textures)
```
